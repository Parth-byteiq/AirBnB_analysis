```python
# here we got to see that there are a lot of lookalikes for the city Austin, lets fix it using fuzzywuzzy library
```

```python
cities = old_listings.city.unique()
```

```python
cities.sort()
```

```python
cities
```

```python
import fuzzywuzzy
from fuzzywuzzy import process
```

```python
old_listings.city = old_listings.city.str.lower()
```

```python
old_listings.city.str.strip()
```

```python
matches = fuzzywuzzy.process.extract('Austin', cities, limit = 26, scorer = fuzzywuzzy.fuzz.token_sort_ratio)
```

```python
matches
```

```python
# as we can see 44 is the least matched score for Austin city
```

```python
def replace_matches_in_col(df, col, string_to_match, min_ratio = 44):
    strings = df[col].unique()
    matches = fuzzywuzzy.process.extract(string_to_match, strings, limit = 20, scorer = fuzzywuzzy.fuzz.token_sort_ratio)
    close_matches = [match[0] for match in matches if match[1]>=min_ratio]
    matching_rows = df[col].isin(close_matches)
    df.loc[matching_rows, col] = string_to_match
    print("Success")
```

```python
replace_matches_in_col(old_listings, 'city','Austin')
```
