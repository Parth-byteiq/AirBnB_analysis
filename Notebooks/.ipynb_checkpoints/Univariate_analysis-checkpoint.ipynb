```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nums = df['neighbourhood'].value_counts()
nums.shape[0]
```

```python
plt.figure(figsize=(12,12))
ax = sns.barplot(x=nums.index, y=nums.values )
plt.xticks(rotation=90)
plt.title('Value counts of neighbourhood')
plt.ylabel('Counts')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=7, color='black', xytext=(0, 5), textcoords='offset points', rotation=90)
plt.savefig("bar_plot.png")
plt.show()
```

```python
import pandas as pd

def parse_info(df):
        d1 = {'Column': [i for i in df.columns],
              'Non-Null count':[str(df.shape[0]-df[i].isnull().sum())+' non-null' for i in df.columns],
              'Dtypes': ['Object' if df[i].dtypes == 'O' else df[i].dtypes for i in df.columns],
              'col_count': df.shape[1],
              'row_count': df.shape[0]
             }
        df_info = pd.DataFrame(d1)
        return df_info
```

```python
parse_info(df)
```
